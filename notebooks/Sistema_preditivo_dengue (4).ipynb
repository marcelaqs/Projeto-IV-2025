# **Referencial Teórico**

As arboviroses, como dengue, chikungunya e Zika, estão entre os principais problemas de saúde pública no Brasil. Essas doenças são transmitidas por vetores e apresentam ciclos epidêmicos influenciados por fatores climáticos, ambientais e socioeconômicos, além do processo acelerado de urbanização (BRASIL, 2023; PAHO, 2023). A previsão de surtos exige, portanto, metodologias capazes de lidar com dados temporais e multivariados, contemplando variáveis epidemiológicas, meteorológicas e demográficas.

A análise de séries temporais constitui uma das principais estratégias para compreender fenômenos que evoluem no tempo. Seus elementos essenciais incluem tendência, sazonalidade, ciclos e variações aleatórias, componentes que permitem a identificação de padrões recorrentes e subsidiam a construção de modelos preditivos (SILVA; PERES; BOSCARIOLI, 2016; SOUZA, 2021). Dentre os métodos estatísticos tradicionais, destacam-se os modelos ARIMA e SARIMA, amplamente aplicados à previsão a partir de dados históricos e reconhecidos por sua capacidade de modelar sazonalidade e dependência temporal (IBM, 2024).

No contexto brasileiro, iniciativas como o InfoDengue têm demonstrado o potencial da integração entre dados epidemiológicos, climáticos e de mobilidade para monitoramento em tempo quase real, oferecendo indicadores de transmissão e risco (LOURENÇO et al., 2020). Esses sistemas evidenciam a importância da combinação entre técnicas estatísticas e ferramentas computacionais mais avançadas para aprimorar a vigilância epidemiológica.

Nos últimos anos, o aprendizado de máquina consolidou-se como alternativa promissora aos modelos tradicionais. Algoritmos como Redes Neurais Recorrentes (RNN) e suas variantes, como a LSTM, destacam-se pela capacidade de capturar dependências de longo prazo e relações não lineares em séries temporais complexas, especialmente em bases multivariadas (AWS, 2024). Embora apresentem desempenho superior em muitos cenários, esses métodos demandam maior volume de dados, capacidade computacional elevada e apresentam desafios relacionados à interpretabilidade, o que pode limitar sua adoção em contextos operacionais.

Dessa forma, tanto os modelos estatísticos quanto os modelos baseados em aprendizado de máquina possuem vantagens e limitações complementares. Modelos estatísticos são mais transparentes, requerem menor quantidade de dados e são adequados para séries mais curtas, enquanto algoritmos de aprendizado de máquina oferecem maior flexibilidade e desempenho em bases extensas e heterogêneas, ainda que com custo computacional superior. Este projeto considera ambas as abordagens de forma integrada, buscando identificar qual delas apresenta melhor aplicabilidade para a previsão de surtos de arboviroses em ambientes urbanos.

### 2. Limpeza e Padronização

Nesta etapa, são realizadas operações fundamentais para garantir a qualidade da base:

- conversão da variável `data_iniSE` para formato datetime;
- ordenação cronológica;
- interpolação linear de valores ausentes em variáveis numéricas;
- preenchimento por média em atributos com ausência sistemática (como `tweet`);
- remoção de colunas totalmente nulas ou irrelevantes.

Essas ações asseguram consistência estrutural e evitam ruídos que possam prejudicar o processo de aprendizado.

Essa etapa garante a integridade do índice temporal, essencial para modelagem baseada em dependência histórica.

```python
# ============================================================
# 1. IMPORTAÇÃO DE BIBLIOTECAS
# ============================================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# ============================================================
# 2. CARREGAMENTO DOS DADOS
# ============================================================
csv_url = 'https://raw.githubusercontent.com/CamillaSemenzzatt/Projeto-IV-/main/dengue_1-49.csv' # Updated URL based on user input
df = pd.read_csv(csv_url)

print("Informações iniciais do DataFrame:")
df.info()

print("\nPrimeiras linhas:")
display(df.head())

# ============================================================
# 3. PRÉ-PROCESSAMENTO INICIAL
# ============================================================
# Conversão da coluna de datas
df['data_iniSE'] = pd.to_datetime(df['data_iniSE'])

# Intervalos de datas e semanas
print(f"\nIntervalo de Datas: {df['data_iniSE'].min().date()} a {df['data_iniSE'].max().date()}")
print(f"Intervalo de Semanas Epidemiológicas: {df['SE'].min()} a {df['SE'].max()}")

# Remoção de colunas irrelevantes
cols_to_drop = ['casprov_est', 'casprov_est_min', 'casprov_est_max', 'casconf']
df.drop(columns=cols_to_drop, inplace=True)
print(f"\nColunas removidas: {cols_to_drop}")

display(df.head())

# ============================================================
# 4. TRATAMENTO DE VALORES AUSENTES
# ============================================================
df['casos_est_max'] = df['casos_est_max'].interpolate(method='linear')
df['tweet'] = df['tweet'].fillna(df['tweet'].mean())

print("\nValores ausentes após tratamento:")
print(df.isnull().sum())

# ============================================================
# 5. INDEXAÇÃO TEMPORAL
# ============================================================
df = df.set_index('data_iniSE').sort_index()

print("\nApós definição do índice temporal:")
display(df.head())
df.info()

# ============================================================
# 6. TRATAMENTO DE DATAS DUPLICADAS
# ============================================================
if df.index.duplicated().any():
    print("\n⚠ Datas duplicadas detectadas. Aplicando agregação...")

    agg_rules = {
        col: ('sum' if col == 'casos' else 'mean')
        if pd.api.types.is_numeric_dtype(df[col]) else 'first'
        for col in df.columns
    }

    df = df.groupby(df.index).agg(agg_rules)

print("\nApós tratamento de duplicações:")
df.info()

# ============================================================
# 7. GERAÇÃO DE VARIÁVEIS COM LAG
# ============================================================
targets = ['casos', 'casos_est']
num_lags = 3

for tgt in targets:
    for lag in range(1, num_lags + 1):
        df[f'{tgt}_lag_{lag}'] = df[tgt].shift(lag)

print("\nDataFrame com variáveis defasadas (lags):")
display(df.head())

# ============================================================
# 8. CÁLCULO DE MÉDIAS MÓVEIS
# ============================================================
windows = [7, 14, 21]

for tgt in targets:
    for w in windows:
        df[f'{tgt}_ma_{w}'] = df[tgt].rolling(window=w).mean()

print("\nDataFrame com médias móveis:")
display(df.head(10))

# ============================================================
# 9. FEATURE ENGINEERING — AMPLITUDE TÉRMICA
# ============================================================
df['amplitude_termica'] = df['tempmax'] - df['tempmin']

print("\nDataFrame com 'amplitude_termica':")
display(df.head())

# ============================================================
# 10. VISUALIZAÇÃO DAS SÉRIES TEMPORAIS
# ============================================================
plt.figure(figsize=(15, 10))

plt.subplot(2, 1, 1)
sns.lineplot(x=df.index, y=df['casos_est'])
plt.title("Série Temporal - Casos Estimados")
plt.grid(True)

plt.subplot(2, 1, 2)
sns.lineplot(x=df.index, y=df['Rt'], color='red')
plt.title("Série Temporal - Rt")
plt.grid(True)

plt.tight_layout()
plt.show()

# ============================================================
# 11. DECOMPOSIÇÃO SAZONAL
# ============================================================
period = 52  # periodicidade semanal

# Casos Estimados
decomp_casos = seasonal_decompose(df['casos_est'], model='additive', period=period)
fig = decomp_casos.plot()
fig.set_size_inches(12, 8)
fig.suptitle("Decomposição Sazonal — Casos Estimados", y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

# Rt
decomp_rt = seasonal_decompose(df['Rt'], model='additive', period=period)
fig = decomp_rt.plot()
fig.set_size_inches(12, 8)
fig.suptitle("Decomposição Sazonal — Rt", y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

# ============================================================
# 12. MATRIZ DE CORRELAÇÃO CLIMÁTICA
# ============================================================
cols_corr = ['casos_est', 'tempmed', 'umidmed', 'amplitude_termica']
corr_matrix = df[cols_corr].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação — Casos Estimados e Variáveis Climáticas')
plt.show()

# ============================================================
# 13. ACF E PACF
# ============================================================
# Casos Estimados
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(df['casos_est'].dropna(), ax=axes[0], lags=50)
axes[0].set_title('Autocorrelação — Casos Estimados')

plot_pacf(df['casos_est'].dropna(), ax=axes[1], lags=50, method='ywm')
axes[1].set_title('Autocorrelação Parcial — Casos Estimados')
plt.tight_layout()
plt.show()

# Rt
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(df['Rt'].dropna(), ax=axes[0], lags=50)
axes[0].set_title('Autocorrelação — Rt')

plot_pacf(df['Rt'].dropna(), ax=axes[1], lags=50, method='ywm')
axes[1].set_title('Autocorrelação Parcial — Rt')
plt.tight_layout()
plt.show()
```

## Modelos

A etapa de modelagem foi desenvolvida utilizando redes neurais recorrentes do tipo GRU (Gated Recurrent Units) com o objetivo de prever, semanalmente, o número de casos estimados de dengue (*casos_est*). A implementação seguiu integralmente o pipeline definido nas etapas anteriores, incluindo preparação da série temporal, engenharia de atributos e organização dos dados em janelas deslizantes para previsão de um passo à frente.

Inicialmente, os dados foram carregados e processados pela função `load_and_preprocess`, responsável pela leitura da base original em formato CSV, conversão da coluna de datas para o tipo *datetime*, ordenação cronológica, remoção de colunas não utilizadas e interpolação linear para preenchimento de valores ausentes. Em seguida, a função `add_features` enriqueceu o conjunto de dados com atributos essenciais para modelagem temporal: lags de até doze semanas, médias móveis com múltiplas janelas (7, 14, 21, 30 e 52 semanas), amplitude térmica, variáveis climáticas disponíveis e codificação senoidal da sazonalidade mensal e semanal. O alvo foi transformado para escala logarítmica (`target_log`) a fim de reduzir assimetria e estabilizar a variância da série.

A organização dos dados para treinamento foi realizada pela função `create_windows`, que estruturou a série temporal no formato *sliding window*, utilizando 52 semanas como entrada e a semana subsequente como saída (`n_steps_out = 1`). A divisão entre treino e teste respeitou a ordem temporal, adotando 85% da série para treinamento e 15% para teste, de forma a evitar vazamento de informação.

Antes do treinamento, tanto as variáveis explicativas quanto o alvo (em escala logarítmica) foram padronizados por `StandardScaler`, garantindo maior estabilidade durante o processo de otimização. O modelo final adotado foi uma GRU simples, composta por:

- uma camada GRU com número ajustável de unidades;
- uma camada de *Dropout* para redução de sobreajuste;
- uma camada densa intermediária com ativação ReLU;
- uma camada de saída linear para previsão contínua.

A arquitetura utilizada é encapsulada na função `build_gru_default`, permitindo execução estável sem necessidade de tuning adicional. O treinamento foi conduzido pela função `train_model`, que emprega os seguintes mecanismos:

- **EarlyStopping**, interrompendo o treinamento quando a validação deixa de melhorar;
- **ReduceLROnPlateau**, reduzindo a taxa de aprendizado automaticamente em caso de estagnação;
- **ModelCheckpoint**, salvando somente o melhor modelo obtido.

A validação utilizou diretamente o conjunto de teste para garantir comparabilidade entre épocas e estabilidade no processo.

O fluxo completo de cada execução é realizado pela função `run_experiment_for_window`, que engloba:

1. criação das janelas temporais;  
2. divisão treino–teste;  
3. normalização dos dados;  
4. treinamento do modelo GRU;  
5. previsão em escala logarítmica;  
6. retorno das previsões à escala original;  
7. cálculo das métricas de erro (MSE, RMSE, MAE);  
8. exportação dos resultados e previsões em formato CSV.

Esse pipeline garante reprodutibilidade, consistência metodológica e integração direta com todas as etapas do projeto, resultando em um modelo capaz de capturar dependências temporais relevantes e produzir previsões confiáveis dos casos estimados de dengue.

```python
import os
import random
import time
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, GRU, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

import keras_tuner as kt

# =========================
# Configuração geral
# =========================
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)

csv_url = "https://raw.githubusercontent.com/CamillaSemenzzatt/Projeto-IV-/main/dengue_1-49.csv"
date_col = "data_iniSE"
target_col = "casos_est"

MODEL_DIR = "models_all"
os.makedirs(MODEL_DIR, exist_ok=True)

WINDOWS = [52]          # janela de 52 semanas
TRAIN_FRAC = 0.85       # melhor split que você usou
EPOCHS = 40
BATCH_SIZE = 32

RUN_TUNER_GRU = False   # para resultado estável no artigo
TUNER_TRIALS = 20
TUNER_EPOCHS = 50


# =========================
# Pré-processamento
# =========================
def load_and_preprocess(csv_url, date_col="data_iniSE", target_col="casos_est"):
    df = pd.read_csv(csv_url, engine='python', on_bad_lines='skip')
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.sort_values(date_col).reset_index(drop=True)

    drop_cols = ["casprov_est", "casprov_est_min", "casprov_est_max", "casconf"]
    df = df.drop(columns=[c for c in drop_cols if c in df.columns], errors="ignore")

    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    df[num_cols] = df[num_cols].interpolate(method="linear", limit_direction="both")

    if "tweet" in df.columns:
        df["tweet"] = df["tweet"].fillna(df["tweet"].mean())

    df.set_index(date_col, inplace=True)
    df.sort_index(inplace=True)
    return df


def add_features(df, target_col="casos_est", max_lag=12, ma_windows=[7, 14, 21, 30, 52]):
    # Lags
    for lag in range(1, max_lag + 1):
        df[f"{target_col}_lag_{lag}"] = df[target_col].shift(lag)

    # Médias móveis
    for w in ma_windows:
        df[f"{target_col}_ma_{w}"] = df[target_col].rolling(window=w, min_periods=1).mean()

    # Amplitude térmica
    if "tempmax" in df.columns and "tempmin" in df.columns:
        df["amplitude_termica"] = df["tempmax"] - df["tempmin"]

    # Sazonalidade
    df["month"] = df.index.month
    df["weekofyear"] = df.index.isocalendar().week.astype(int)

    df["month_sin"] = np.sin(2 * np.pi * df["month"] / 12)
    df["month_cos"] = np.cos(2 * np.pi * df["month"] / 12)
    df["week_sin"] = np.sin(2 * np.pi * df["weekofyear"] / 52)
    df["week_cos"] = np.cos(2 * np.pi * df["weekofyear"] / 52)

    # Alvo em log
    df["target_log"] = np.log1p(df[target_col])

    df.dropna(inplace=True)
    return df


def create_windows(df, feature_cols, target_col_log="target_log",
                   n_steps_in=52, n_steps_out=1):
    arr = df[feature_cols].values
    t = df[target_col_log].values

    X, y = [], []
    max_start = len(df) - (n_steps_in + n_steps_out) + 1

    for start in range(max_start):
        end_x = start + n_steps_in
        end_y = end_x + n_steps_out
        X.append(arr[start:end_x, :])
        y.append(t[end_x:end_y])

    X = np.array(X)
    y = np.array(y)

    if n_steps_out == 1:
        y = y.reshape(-1)

    return X, y


def train_test_split_time(X, y, train_frac=0.85):
    split = int(len(X) * train_frac)
    return X[:split], X[split:], y[:split], y[split:]


def scale_train_test(X_train, X_test, y_train, y_test):
    """
    Escala features e alvo EM ESCALA LOG (igual no melhor modelo).
    """
    ns, ts, nf = X_train.shape

    scaler_X = StandardScaler().fit(X_train.reshape(-1, nf))
    X_train_s = scaler_X.transform(X_train.reshape(-1, nf)).reshape(ns, ts, nf)
    X_test_s = scaler_X.transform(X_test.reshape(-1, nf)).reshape(X_test.shape[0], ts, nf)

    scaler_y = StandardScaler().fit(y_train.reshape(-1, 1))
    y_train_s = scaler_y.transform(y_train.reshape(-1, 1)).reshape(-1)
    y_test_s = scaler_y.transform(y_test.reshape(-1, 1)).reshape(-1)

    return X_train_s, X_test_s, y_train_s, y_test_s, scaler_X, scaler_y


def metrics_original(y_true_log, y_pred_log):
    """
    Recebe verdadeiro e previsto em LOG (sem padronização),
    converte para casos reais e calcula métricas.
    """
    y_true = np.expm1(y_true_log)
    y_pred = np.expm1(y_pred_log)

    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    return mse, rmse, mae, y_true, y_pred


# =========================
# Modelo GRU (simples, o que funcionou melhor)
# =========================
def build_gru_tunable(hp, n_timesteps, n_features):
    units1 = hp.Int("gru_units1", 32, 256, step=32, default=128)
    dropout = hp.Float("dropout", 0.0, 0.4, step=0.05, default=0.15)
    lr = hp.Choice("lr", [1e-2, 1e-3, 1e-4], default=1e-3)

    model = Sequential([
        Input(shape=(n_timesteps, n_features)),
        GRU(units1, return_sequences=False),
        Dropout(dropout),
        Dense(hp.Int("dense", 16, 64, step=16, default=32), activation="relu"),
        Dense(1)
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss="mse"
    )
    return model


def build_gru_default(n_timesteps, n_features):
    hp = kt.HyperParameters()
    return build_gru_tunable(hp, n_timesteps, n_features)


# =========================
# Treino
# =========================
def train_model(model, X_train, y_train, X_val, y_val, model_path,
                epochs=EPOCHS, batch_size=BATCH_SIZE):

    es = EarlyStopping(
        monitor="val_loss",
        patience=20,
        restore_best_weights=True,
        verbose=1
    )
    rlr = ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=8,
        min_lr=1e-5,
        verbose=1
    )
    ckpt = ModelCheckpoint(
        model_path,
        monitor="val_loss",
        save_best_only=True,
        verbose=1
    )

    hist = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),   # usando teste como validação
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[es, rlr, ckpt],
        verbose=2,
        shuffle=False
    )
    return hist


def run_experiment_for_window(df, feature_cols, window,
                              run_tuner_gru=RUN_TUNER_GRU,
                              tuner_trials=TUNER_TRIALS):
    print("\n==========================")
    print("WINDOW:", window)
    print("========================")

    X, y = create_windows(
        df,
        feature_cols,
        target_col_log="target_log",
        n_steps_in=window
    )
    print("Total samples (after windows):", len(X))

    X_train, X_test, y_train, y_test = train_test_split_time(X, y, TRAIN_FRAC)

    X_train_s, X_test_s, y_train_s, y_test_s, scaler_X, scaler_y = scale_train_test(
        X_train, X_test, y_train, y_test
    )

    n_timesteps, n_features = X_train_s.shape[1], X_train_s.shape[2]
    results_local = {}

    # ---------- GRU ----------
    if run_tuner_gru:
        print("Rodando KerasTuner GRU")
        tuner_gru = kt.RandomSearch(
            lambda hp: build_gru_tunable(hp, n_timesteps, n_features),
            objective="val_loss",
            max_trials=tuner_trials,
            executions_per_trial=1,
            directory="kt_gru",
            project_name=f"gru_w{window}"
        )
        tuner_gru.search(
            X_train_s, y_train_s,
            epochs=TUNER_EPOCHS,
            validation_data=(X_test_s, y_test_s),
            shuffle=False,
            verbose=1
        )
        best_hp_gru = tuner_gru.get_best_hyperparameters(1)[0]
        model_gru = build_gru_tunable(best_hp_gru, n_timesteps, n_features)
    else:
        model_gru = build_gru_default(n_timesteps, n_features)

    path_gru = os.path.join(MODEL_DIR, f"gru_w{window}.keras")
    train_model(model_gru, X_train_s, y_train_s, X_test_s, y_test_s, path_gru)

    # Previsão no conjunto de teste (EM ESCALA PADRONIZADA DE LOG)
    y_pred_gru_s = model_gru.predict(X_test_s, verbose=0).reshape(-1, 1)
    y_pred_gru_log = scaler_y.inverse_transform(y_pred_gru_s).reshape(-1)

    # Volta o y_test para a escala de log original
    y_test_log = scaler_y.inverse_transform(y_test_s.reshape(-1, 1)).reshape(-1)

    mse_g, rmse_g, mae_g, y_true_g, y_pred_g = metrics_original(
        y_test_log, y_pred_gru_log
    )
    results_local["GRU"] = {
        "mse": mse_g, "rmse": rmse_g, "mae": mae_g,
        "y_true": y_true_g, "y_pred": y_pred_g
    }

    # ---------- Salva previsões ----------
    idx = df.index[-len(y_true_g):]
    outdf = pd.DataFrame(
        {
            "y_true": y_true_g,
            "pred_gru": y_pred_g,
        },
        index=idx
    )
    outdf.to_csv(os.path.join(MODEL_DIR, f"preds_w{window}.csv"))

    return results_local


# =========================
# Execução
# =========================
df_raw = load_and_preprocess(csv_url, date_col=date_col, target_col=target_col)
df_feat = add_features(
    df_raw,
    target_col=target_col,
    max_lag=12,
    ma_windows=[7, 14, 21, 30, 52]
)

climatic_cols = [
    c for c in ["tempmed", "umidmed", "precip", "tweet", "Rt"]
    if c in df_feat.columns
]

feature_cols = [f"{target_col}_lag_{lag}" for lag in range(1, 13)]
feature_cols += [f"{target_col}_ma_{w}" for w in [7, 14, 21, 30, 52]]
feature_cols += ["amplitude_termica", "month_sin", "month_cos", "week_sin", "week_cos"]
feature_cols += climatic_cols
feature_cols = [c for c in feature_cols if c in df_feat.columns]

print("Features finais:", feature_cols)

all_results = {}
start_total = time.time()

for w in WINDOWS:
    res = run_experiment_for_window(
        df_feat,
        feature_cols,
        w,
        run_tuner_gru=RUN_TUNER_GRU,
        tuner_trials=TUNER_TRIALS
    )
    all_results[w] = res

end_total = time.time()
print("Tempo total (s):", end_total - start_total)

rows = []
for w, r in all_results.items():
    for mname, metrics in r.items():
        rows.append(
            {
                "window": w,
                "model": mname,
                "mse": metrics["mse"],
                "rmse": metrics["rmse"],
                "mae": metrics["mae"]
            }
        )

res_df = pd.DataFrame(rows).sort_values(["window", "rmse"]).reset_index(drop=True)
print(res_df)

last_w = WINDOWS[-1]
preds_path = os.path.join(MODEL_DIR, f"preds_w{last_w}.csv")

if os.path.exists(preds_path):
    preds = pd.read_csv(preds_path, index_col=0, parse_dates=True)
    plt.figure(figsize=(12, 4))
    plt.plot(preds["y_true"], label="Real")
    plt.plot(preds["pred_gru"], label="GRU")
    plt.legend()
    plt.title(f"Predições (window={last_w})")
    plt.tight_layout()
    plt.show()

summary_path = os.path.join(MODEL_DIR, "summary_results.csv")
res_df.to_csv(summary_path, index=False)
print("Resumo salvo em:", summary_path)
```
