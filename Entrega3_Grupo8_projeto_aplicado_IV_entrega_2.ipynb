# **Introdução**

As arboviroses — como dengue e chikungunya — configuram-se como um dos principais desafios para a saúde pública no Brasil, especialmente em áreas urbanas com alta densidade populacional e infraestrutura sanitária precária. A combinação entre fatores climáticos, urbanização acelerada e ausência de políticas preventivas eficazes favorece a proliferação do mosquito vetor. Esse quadro se insere no contexto mais amplo das Doenças Tropicais Negligenciadas (DTNs), que afetam de forma desproporcional populações em situação de vulnerabilidade socioeconômica. Apesar de causarem elevada morbidade e mortalidade, as DTNs historicamente recebem menor prioridade em investimentos e políticas de pesquisa, em parte pelo baixo retorno econômico que oferecem à indústria farmacêutica (HOTEZ et al., 2006; BRASIL, 2023). Embora o país disponha de sistemas consolidados de notificação e monitoramento, a resposta aos surtos ainda é majoritariamente reativa, com medidas implementadas apenas após o aumento expressivo dos casos. Essa lacuna entre a disponibilidade de dados e sua aplicação efetiva no planejamento de ações preventivas evidencia a importância de ferramentas capazes de transformar informação em suporte concreto à tomada de decisão, favorecendo intervenções mais ágeis e eficazes, sobretudo em áreas urbanas vulneráveis.

# **Referencial Teórico**

As arboviroses — como dengue, chikungunya e Zika — configuram-se entre os principais problemas de saúde pública no Brasil, caracterizando-se pela transmissão vetorial e por ciclos epidêmicos influenciados tanto pela sazonalidade climática quanto pela urbanização acelerada (BRASIL, 2023; PAHO, 2023). A previsão de surtos dessas doenças exige métodos capazes de lidar com dados temporais e multivariados, considerando variáveis epidemiológicas, ambientais e sociodemográficas.

A análise de séries temporais é uma das principais abordagens para compreender fenômenos ao longo do tempo. Seus componentes básicos incluem tendência, sazonalidade, ciclos e ruído aleatório, os quais permitem identificar padrões recorrentes e apoiar a modelagem preditiva (SILVA; PERES; BOSCARIOLI, 2016; SOUZA, 2021). Entre os modelos estatísticos clássicos mais utilizados destacam-se o ARIMA e o SARIMA, amplamente empregados na previsão de eventos a partir de dados históricos (IBM, 2024).

No cenário brasileiro, iniciativas como o InfoDengue têm se consolidado como ferramentas de monitoramento e previsão de arboviroses, integrando dados epidemiológicos, climáticos e de mobilidade em tempo quase real (LOURENÇO et al., 2020). Esses sistemas demonstram a relevância de combinar métodos estatísticos e técnicas de aprendizado de máquina para fortalecer a vigilância epidemiológica.

Nos últimos anos, o aprendizado de máquina tem ganhado destaque como alternativa aos modelos tradicionais. Algoritmos como Redes Neurais Recorrentes (RNN) e suas variantes, como a LSTM, apresentam maior capacidade para capturar relações complexas e não lineares em séries temporais, oferecendo bom desempenho em cenários com múltiplas variáveis preditoras (AWS, 2024). No entanto, esses modelos demandam grandes volumes de dados, maior poder computacional e apresentam limitações relacionadas à interpretabilidade.

Assim, tanto os modelos estatísticos quanto os de aprendizado de máquina possuem vantagens e limitações. Enquanto os primeiros são mais simples, transparentes e adequados para séries curtas, os segundos alcançam melhor desempenho em bases de dados extensas e heterogêneas, ainda que a um custo computacional mais elevado. Este projeto propõe explorar essas abordagens de forma complementar, avaliando sua aplicabilidade para a predição de surtos de arboviroses em áreas urbanas.

# **Pipeline da Solução**

O pipeline a seguir descreve as etapas para processar os dados e chegar na solução preditiva, considerando **dados semanais entre 2020 e 2024**.

---

## 1. Coleta dos Dados
- **Epidemiológicos**: Sregistros do SINAN (casos notificados de arboviroses) e informações complementares do InfoDengue.
- **Climáticos**: dados do INMET e CPTEC (temperatura, umidade, precipitação).  
- **Período**: 2020–2024, com granularidade semanal.  

---

## 2. Pré-processamento
- Padronização de datas e unidades geográficas (município, estado).
- Tratamento de valores faltantes e inconsistentes (NA, “ignorado”).
- - Integração dos bancos de dados epidemiológicos e climáticos em uma única base consolidada.

---

## 3. Análise Exploratória
- Visualização das séries temporais por meio de gráficos de linha, heatmaps e correlogramas.
- dentificação dos componentes das séries: **tendência, sazonalidade, ciclos e ruído**.
- Testes estatísticos de estacionaridade (ex.: Dickey-Fuller aumentado).  

---

## 4. Transformação dos Dados
- Criação de variáveis derivadas (lags, médias móveis, índices climáticos).
- Normalização/padronização das variáveis para garantir comparabilidade.
- Seleção de atributos relevantes para modelagem.

---

## 5. Modelagem Preditiva
- **Modelos estatísticos**: ARIMA, SARIMA, Prophet.  
- **Modelos de aprendizado de máquina**: Redes Neurais Recorrentes (RNN) e LSTM.
- Ajuste de hiperparâmetros e comparação de desempenho entre abordagens.

---

##  6. Validação
- Divisão treino (2020–2023) e teste (2024).  
- Avaliação com métricas: MAE, RMSE, acurácia.  
- Comparação entre modelos estatísticos e de aprendizado de máquina.

---

## 7. Aplicação e Visualização
- Geração de alertas semanais de risco preditivo.
- Construção de dashboard interativo para gestores de saúde, com indicadores como casos esperados vs. observados.
- Apoio direto ao planejamento de campanhas preventivas e políticas públicas.

---

**Fluxo Resumido:**  
`Coleta → Pré-processamento → Exploração → Transformação → Modelagem → Validação → Aplicação`

```python
# === CONCATENAÇÃO EM STREAMING COM ESQUEMA FIXO (ROBUSTO ENTRE ANOS) ===
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
from pathlib import Path

CSV_BY_YEAR = {
    2022: RAW_DIR / "DENGBR22.csv",
    2023: RAW_DIR / "DENGBR23.csv",
    2024: RAW_DIR / "DENGBR24.csv",
}

PARQUET_UNICO = PROC_DIR / "dengue_2022_2024_bruto.parquet"
CHUNK = 200_000  # pode reduzir para 100_000 se quiser usar menos RAM

# 1) Descubro o conjunto-unificado de colunas (apenas cabeçalho, sem carregar dados)
cols_union = []
seen = set()
for yr, path in CSV_BY_YEAR.items():
    if not path.exists():
        raise FileNotFoundError(f"Arquivo não encontrado: {path}")
    hdr = pd.read_csv(path, sep=",", encoding="utf-8", nrows=0)
    for c in hdr.columns.tolist():
        if c not in seen:
            seen.add(c)
            cols_union.append(c)

# acrescento 'ano_origem' ao final (não existe no bruto)
if "ano_origem" not in cols_union:
    cols_union.append("ano_origem")

# 2) Defino um schema fixo: todas as colunas como string, exceto 'ano_origem' (int64)
fields = []
for c in cols_union:
    if c == "ano_origem":
        fields.append(pa.field(c, pa.int64()))
    else:
        fields.append(pa.field(c, pa.string()))
schema = pa.schema(fields)

# Se já existe, apago para reescrever do zero
if PARQUET_UNICO.exists():
    PARQUET_UNICO.unlink()

writer = None
linhas_por_ano = {yr: 0 for yr in CSV_BY_YEAR}

def _coerce_chunk_to_schema(df_chunk: pd.DataFrame) -> pa.Table:
    """
    - Garante TODAS as colunas (reindex no superset).
    - Converte todas as colunas (exceto ano_origem) para pandas StringDtype().
    - ano_origem -> Int64 (pandas) para ir como int64 no Arrow.
    - Converte para Arrow Table usando o schema fixo.
    """
    # garante superset (colunas que não existem viram NA)
    df_chunk = df_chunk.reindex(columns=cols_union, fill_value=pd.NA)

    # tipos pandas estáveis
    for c in cols_union:
        if c == "ano_origem":
            # garante numérico inteiro com NA seguro
            df_chunk[c] = pd.to_numeric(df_chunk[c], errors="coerce").astype("Int64")
        else:
            # tudo como string (mesmo que a coluna esteja toda vazia)
            if df_chunk[c].dtype.name != "string":
                df_chunk[c] = df_chunk[c].astype("string")

    # converto para Arrow com o schema alvo (evita colunas 'null')
    table = pa.Table.from_pandas(df_chunk, schema=schema, preserve_index=False)
    return table

# 3) Escrevo em streaming com schema fixo
for yr, csv_path in CSV_BY_YEAR.items():
    print(f"→ Processando {csv_path.name} em chunks de {CHUNK:,} linhas...")
    for i, chunk in enumerate(pd.read_csv(csv_path, sep=",", encoding="utf-8",
                                          dtype=str, low_memory=False, chunksize=CHUNK)):
        # adiciono a coluna do ano de origem
        chunk["ano_origem"] = yr
        linhas_por_ano[yr] += len(chunk)

        table = _coerce_chunk_to_schema(chunk)

        if writer is None:
            writer = pq.ParquetWriter(PARQUET_UNICO, schema, compression="snappy")

        writer.write_table(table)

        if (i + 1) % 5 == 0:
            print(f"   · {csv_path.name}: {((i+1)*CHUNK):,} linhas já escritas...")

if writer is not None:
    writer.close()

print("✓ Parquet único salvo em:", PARQUET_UNICO)
print("Linhas por ano:", {k: f'{v:,}' for k, v in linhas_por_ano.items()})
print("Total de linhas:", f"{sum(linhas_por_ano.values()):,}")

```

```python
# === INSPEÇÃO LEVE DO PARQUET (HEAD + CONTAGEM POR ano_origem) ===
import pyarrow.parquet as pq


PARQUET_UNICO = PROC_DIR / "dengue_2022_2024_bruto.parquet"
pf = pq.ParquetFile(PARQUET_UNICO)

# 1) HEAD seguro (~5 linhas) lendo o primeiro(s) row group(s)
n_quero = 5
coletados, lidos = [], 0
for rg in range(pf.num_row_groups):
    df_rg = pf.read_row_group(rg).to_pandas()
    coletados.append(df_rg)
    lidos += len(df_rg)
    if lidos >= n_quero:
        break
df_head = pd.concat(coletados, ignore_index=True).head(n_quero)
print("Head do parquet concatenado:")
display(df_head)

# 2) Contagem por ano_origem (lendo só a coluna, super leve)
anos = pf.read(columns=["ano_origem"]).to_pandas()
contagem_ano_origem = anos["ano_origem"].value_counts().sort_index()
print("\nNotificações por ano_origem (do parquet):")
display(contagem_ano_origem)
```

#Pré-processamento

```python

# ------------------------------------------------------------
# 3) PADRONIZAR NOMES E TIPOS BÁSICOS
# ------------------------------------------------------------
# Padronizar nomes em MAIÚSCULAS (mantemos padrão SINAN)
df.columns = [c.strip().upper() for c in df.columns]

# Converter datas (se existirem)
for col in ["DT_NOTIFIC", "DT_SIN_PRI", "DT_ENCERRA"]:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
```

```python
# ------------------------------------------------------------
# 4) IDADE EM ANOS + ATRASO SINTOMAS→NOTIFICAÇÃO
# ------------------------------------------------------------
# NU_IDADE_N costuma estar em anos neste recorte — garantimos plausibilidade
if "NU_IDADE_N" in df.columns:
    def _idade_anos(x):
        try:
            v = float(x)
            return v if 0 <= v <= 120 else np.nan
        except:
            return np.nan
    df["IDADE_ANOS"] = df["NU_IDADE_N"].apply(_idade_anos)
else:
    df["IDADE_ANOS"] = np.nan

# Dias entre início dos sintomas e notificação
if {"DT_SIN_PRI", "DT_NOTIFIC"}.issubset(df.columns):
    df["DIAS_ONSET_NOTIF"] = (df["DT_NOTIFIC"] - df["DT_SIN_PRI"]).dt.days
else:
    df["DIAS_ONSET_NOTIF"] = np.nan
```

```python
# ------------------------------------------------------------
# 5) PADRONIZAR SEXO E FAIXAS ETÁRIAS
# ------------------------------------------------------------
if "CS_SEXO" in df.columns:
    # Normaliza para 'Masculino'/'Feminino' e mantém NaN quando não identificado
    s = df["CS_SEXO"].astype(str).str.upper().str[0]

    # Use .loc for conditional assignment to handle mixed types more robustly
    df["CS_SEXO"] = None # Initialize the column with None or NaN
    df.loc[s == "M", "CS_SEXO"] = "Masculino"
    df.loc[s == "F", "CS_SEXO"] = "Feminino"
    # Values where s is neither 'M' nor 'F' will remain None (which pandas often treats as NaN for object dtype)

else:
    df["CS_SEXO"] = np.nan

# Faixas etárias úteis
bins   = [0,5,12,18,30,45,60,75,120]
labels = ["0-4","5-11","12-17","18-29","30-44","45-59","60-74","75+"]
df["FAIXA_ETARIA"] = pd.cut(df["IDADE_ANOS"], bins=bins, labels=labels, right=False)
```

```python
# ------------------------------------------------------------
# 6) CLASSIFICAÇÃO FINAL E EVOLUÇÃO (TARGET CANDIDATO)
# ------------------------------------------------------------
# Mantemos colunas brutas para referência, se existirem
if "CLASSI_FIN" in df.columns:
    df["CLASSIFICACAO_FINAL"] = df["CLASSI_FIN"]

if "EVOLUCAO" in df.columns:
    df["EVOLUCAO_FINAL"] = df["EVOLUCAO"]

# Criação de binário de óbito (ajuste se sua codificação for diferente)
# Convenção comum no SINAN: 1=Cura, 2=Óbito, 9=Ign/Branco
if "EVOLUCAO_FINAL" in df.columns:
    df["EVOLUCAO_STR"] = df["EVOLUCAO_FINAL"].astype(str).str.strip()
    df["EVOLUCAO_BIN"] = np.where(df["EVOLUCAO_STR"]=="2", 1,
                           np.where(df["EVOLUCAO_STR"]=="1", 0, np.nan))
else:
    df["EVOLUCAO_BIN"] = np.nan
```

```python
# ------------------------------------------------------------
# 7) CHECAGENS RÁPIDAS (SANITY CHECK)
# ------------------------------------------------------------
print("\nChecagem rápida:")
print("Qtd datas DT_NOTIFIC válidas:", df["DT_NOTIFIC"].notna().sum() if "DT_NOTIFIC" in df.columns else 0)
print("Qtd datas DT_SIN_PRI válidas:", df["DT_SIN_PRI"].notna().sum() if "DT_SIN_PRI" in df.columns else 0)
print("Faixa etária (top):")
print(df["FAIXA_ETARIA"].value_counts(dropna=False).head())
print("Sexo (top):")
print(df["CS_SEXO"].value_counts(dropna=False).head())
print("Evolução BIN (0=cura,1=óbito):")
print(df["EVOLUCAO_BIN"].value_counts(dropna=False))
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt # Import matplotlib for plotting

serie = serie_casos.set_index('SEMANA')['CASOS']
modelo = SARIMAX(serie, order=(1,1,1), seasonal_order=(1,1,1,52))
ajuste = modelo.fit(disp=False)
previsao = ajuste.forecast(steps=12)

plt.figure(figsize=(12,5))
plt.plot(serie, label='Observado')
plt.plot(previsao, label='Previsto', color='red')
plt.legend()
plt.show()
```
