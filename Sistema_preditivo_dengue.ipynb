# **Referencial Teórico**

As arboviroses, como dengue, chikungunya e Zika, estão entre os principais problemas de saúde pública no Brasil. Essas doenças são transmitidas por vetores e apresentam ciclos epidêmicos influenciados por fatores climáticos, ambientais e socioeconômicos, além do processo acelerado de urbanização (BRASIL, 2023; PAHO, 2023). A previsão de surtos exige, portanto, metodologias capazes de lidar com dados temporais e multivariados, contemplando variáveis epidemiológicas, meteorológicas e demográficas.

A análise de séries temporais constitui uma das principais estratégias para compreender fenômenos que evoluem no tempo. Seus elementos essenciais incluem tendência, sazonalidade, ciclos e variações aleatórias, componentes que permitem a identificação de padrões recorrentes e subsidiam a construção de modelos preditivos (SILVA; PERES; BOSCARIOLI, 2016; SOUZA, 2021). Dentre os métodos estatísticos tradicionais, destacam-se os modelos ARIMA e SARIMA, amplamente aplicados à previsão a partir de dados históricos e reconhecidos por sua capacidade de modelar sazonalidade e dependência temporal (IBM, 2024).

No contexto brasileiro, iniciativas como o InfoDengue têm demonstrado o potencial da integração entre dados epidemiológicos, climáticos e de mobilidade para monitoramento em tempo quase real, oferecendo indicadores de transmissão e risco (LOURENÇO et al., 2020). Esses sistemas evidenciam a importância da combinação entre técnicas estatísticas e ferramentas computacionais mais avançadas para aprimorar a vigilância epidemiológica.

Nos últimos anos, o aprendizado de máquina consolidou-se como alternativa promissora aos modelos tradicionais. Algoritmos como Redes Neurais Recorrentes (RNN) e suas variantes, como a LSTM, destacam-se pela capacidade de capturar dependências de longo prazo e relações não lineares em séries temporais complexas, especialmente em bases multivariadas (AWS, 2024). Embora apresentem desempenho superior em muitos cenários, esses métodos demandam maior volume de dados, capacidade computacional elevada e apresentam desafios relacionados à interpretabilidade, o que pode limitar sua adoção em contextos operacionais.

Dessa forma, tanto os modelos estatísticos quanto os modelos baseados em aprendizado de máquina possuem vantagens e limitações complementares. Modelos estatísticos são mais transparentes, requerem menor quantidade de dados e são adequados para séries mais curtas, enquanto algoritmos de aprendizado de máquina oferecem maior flexibilidade e desempenho em bases extensas e heterogêneas, ainda que com custo computacional superior. Este projeto considera ambas as abordagens de forma integrada, buscando identificar qual delas apresenta melhor aplicabilidade para a previsão de surtos de arboviroses em ambientes urbanos.

7. Preparação Final para Modelagem

Após todo o pré-processamento, o dataset resultante está sem valores ausentes relevantes, organizado em um índice temporal coerente, enriquecido com variáveis derivadas e devidamente padronizado. O conjunto foi preparado para a etapa de criação das janelas temporais (sliding windows) e para o escalonamento das variáveis, constituindo uma base consistente, completa e apropriada para treinar modelos preditivos complexos.

```python
# ============================================================
# 1. IMPORTAÇÃO DE BIBLIOTECAS
# ============================================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# ============================================================
# 2. CARREGAMENTO DOS DADOS
# ============================================================
csv_url = 'https://raw.githubusercontent.com/CamillaSemenzzatt/Projeto-IV-/main/dengue_52-48.csv' # Updated URL based on user input
df = pd.read_csv(csv_url)

print("Informações iniciais do DataFrame:")
df.info()

print("\nPrimeiras linhas:")
display(df.head())

# ============================================================
# 3. PRÉ-PROCESSAMENTO INICIAL
# ============================================================
# Conversão da coluna de datas
df['data_iniSE'] = pd.to_datetime(df['data_iniSE'])

# Intervalos de datas e semanas
print(f"\nIntervalo de Datas: {df['data_iniSE'].min().date()} a {df['data_iniSE'].max().date()}")
print(f"Intervalo de Semanas Epidemiológicas: {df['SE'].min()} a {df['SE'].max()}")

# Remoção de colunas irrelevantes
cols_to_drop = ['casprov_est', 'casprov_est_min', 'casprov_est_max', 'casconf']
df.drop(columns=cols_to_drop, inplace=True)
print(f"\nColunas removidas: {cols_to_drop}")

display(df.head())

# ============================================================
# 4. TRATAMENTO DE VALORES AUSENTES
# ============================================================
df['casos_est_max'] = df['casos_est_max'].interpolate(method='linear')
df['tweet'] = df['tweet'].fillna(df['tweet'].mean())

print("\nValores ausentes após tratamento:")
print(df.isnull().sum())

# ============================================================
# 5. INDEXAÇÃO TEMPORAL
# ============================================================
df = df.set_index('data_iniSE').sort_index()

print("\nApós definição do índice temporal:")
display(df.head())
df.info()

# ============================================================
# 6. TRATAMENTO DE DATAS DUPLICADAS
# ============================================================
if df.index.duplicated().any():
    print("\n⚠ Datas duplicadas detectadas. Aplicando agregação...")

    agg_rules = {
        col: ('sum' if col == 'casos' else 'mean')
        if pd.api.types.is_numeric_dtype(df[col]) else 'first'
        for col in df.columns
    }

    df = df.groupby(df.index).agg(agg_rules)

print("\nApós tratamento de duplicações:")
df.info()

# ============================================================
# 7. GERAÇÃO DE VARIÁVEIS COM LAG
# ============================================================
targets = ['casos', 'casos_est']
num_lags = 3

for tgt in targets:
    for lag in range(1, num_lags + 1):
        df[f'{tgt}_lag_{lag}'] = df[tgt].shift(lag)

print("\nDataFrame com variáveis defasadas (lags):")
display(df.head())

# ============================================================
# 8. CÁLCULO DE MÉDIAS MÓVEIS
# ============================================================
windows = [7, 14, 21]

for tgt in targets:
    for w in windows:
        df[f'{tgt}_ma_{w}'] = df[tgt].rolling(window=w).mean()

print("\nDataFrame com médias móveis:")
display(df.head(10))

# ============================================================
# 9. FEATURE ENGINEERING — AMPLITUDE TÉRMICA
# ============================================================
df['amplitude_termica'] = df['tempmax'] - df['tempmin']

print("\nDataFrame com 'amplitude_termica':")
display(df.head())

# ============================================================
# 10. VISUALIZAÇÃO DAS SÉRIES TEMPORAIS
# ============================================================
plt.figure(figsize=(15, 10))

plt.subplot(2, 1, 1)
sns.lineplot(x=df.index, y=df['casos_est'])
plt.title("Série Temporal - Casos Estimados")
plt.grid(True)

plt.subplot(2, 1, 2)
sns.lineplot(x=df.index, y=df['Rt'], color='red')
plt.title("Série Temporal - Rt")
plt.grid(True)

plt.tight_layout()
plt.show()

# ============================================================
# 11. DECOMPOSIÇÃO SAZONAL
# ============================================================
period = 52  # periodicidade semanal

# Casos Estimados
decomp_casos = seasonal_decompose(df['casos_est'], model='additive', period=period)
fig = decomp_casos.plot()
fig.set_size_inches(12, 8)
fig.suptitle("Decomposição Sazonal — Casos Estimados", y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

# Rt
decomp_rt = seasonal_decompose(df['Rt'], model='additive', period=period)
fig = decomp_rt.plot()
fig.set_size_inches(12, 8)
fig.suptitle("Decomposição Sazonal — Rt", y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

# ============================================================
# 12. MATRIZ DE CORRELAÇÃO CLIMÁTICA
# ============================================================
cols_corr = ['casos_est', 'tempmed', 'umidmed', 'amplitude_termica']
corr_matrix = df[cols_corr].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação — Casos Estimados e Variáveis Climáticas')
plt.show()

# ============================================================
# 13. ACF E PACF
# ============================================================
# Casos Estimados
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(df['casos_est'].dropna(), ax=axes[0], lags=50)
axes[0].set_title('Autocorrelação — Casos Estimados')

plot_pacf(df['casos_est'].dropna(), ax=axes[1], lags=50, method='ywm')
axes[1].set_title('Autocorrelação Parcial — Casos Estimados')
plt.tight_layout()
plt.show()

# Rt
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(df['Rt'].dropna(), ax=axes[0], lags=50)
axes[0].set_title('Autocorrelação — Rt')

plot_pacf(df['Rt'].dropna(), ax=axes[1], lags=50, method='ywm')
axes[1].set_title('Autocorrelação Parcial — Rt')
plt.tight_layout()
plt.show()
```

# Modelos

A etapa de modelagem foi construída para avaliar o desempenho de redes neurais recorrentes aplicadas à previsão semanal de casos estimados de dengue (casos_est), utilizando exclusivamente dados do InfoDengue. A preparação dos dados para os modelos seguiu o pipeline descrito anteriormente: pré-processamento, criação de variáveis derivadas e organização em série temporal.

Primeiro, os dados brutos foram carregados e pré-processados pela função load_and_preprocess, que realiza leitura do CSV, conversão da coluna de datas para o tipo datetime, ordenação cronológica, remoção de colunas inutilizadas e interpolação dos valores numéricos ausentes. Em seguida, a função add_features enriquece a base com atributos específicos para séries temporais: lags do alvo em até 12 defasagens, médias móveis em diferentes janelas, amplitude térmica (tempmax – tempmin) e variáveis cíclicas de mês e semana epidemiológica (seno e cosseno). O alvo é transformado em escala logarítmica (target_log) para estabilizar a variância.

A construção das janelas temporais é feita pela função create_windows, que organiza os dados em blocos do tipo sliding window para previsão de um passo à frente (N_STEPS_OUT = 1). São testados diversos tamanhos de janela (12, 24, 36, 48, 52, 60, 72 e 104 semanas), permitindo comparar o efeito do histórico considerado no desempenho dos modelos. A divisão entre treino e teste respeita a ordem temporal, utilizando aproximadamente 85% da série para treino e 15% para teste, por meio da função train_test_split_time.

Antes do treinamento, as features e o alvo são padronizados com StandardScaler, garantindo maior estabilidade numérica durante a otimização. A modelagem utiliza duas arquiteturas principais de redes neurais recorrentes: GRU e LSTM, ambas empilhadas em duas camadas, com camadas de Dropout intermediárias e uma camada densa final para saída de um valor contínuo. As funções build_gru_tunable e build_lstm_tunable definem essas arquiteturas de forma parametrizável, permitindo que hiperparâmetros como número de unidades, taxa de dropout, tamanho da camada densa e taxa de aprendizado sejam ajustados automaticamente.

O ajuste de hiperparâmetros é realizado com o KerasTuner, usando busca aleatória (RandomSearch) para encontrar as melhores configurações com base na perda de validação (val_loss). O treinamento é conduzido pela função train_model, que aplica callbacks de EarlyStopping, ReduceLROnPlateau e ModelCheckpoint, com o objetivo de evitar overfitting, ajustar dinamicamente a taxa de aprendizado e salvar o melhor modelo encontrado.

Para cada tamanho de janela, a função run_experiment_for_window executa todo o fluxo: criação de janelas, divisão temporal, normalização, ajuste de GRU e LSTM com ou sem tuner, geração de previsões em escala logarítmica, retorno à escala original, cálculo das métricas de erro (MSE, RMSE e MAE) e armazenamento dos resultados. Além dos modelos individuais, são construídos ensembles: um ensemble de média simples das previsões de GRU e LSTM e um ensemble ponderado, em que os pesos são proporcionais ao inverso do RMSE de cada modelo, privilegiando o modelo com melhor desempenho.

Ao final, os resultados para cada combinação de janela e modelo são agregados em um DataFrame (res_df), permitindo comparar desempenho entre modelos e janelas, e são gerados gráficos de valores reais versus previstos para a última janela analisada. Também é salvo um arquivo summary_results.csv com o resumo das métricas, garantindo reprodutibilidade e documentação da etapa de modelagem.

Resultados

Os resultados da modelagem são sintetizados na tabela res_df, que reúne, para cada tamanho de janela temporal testado, as métricas de desempenho (MSE, RMSE e MAE) obtidas pelos modelos GRU, LSTM, ensemble médio e ensemble ponderado. Essa estrutura permite avaliar simultaneamente o impacto do comprimento da janela e do tipo de modelo sobre a qualidade das previsões.

De forma geral, observa-se que o erro tende a variar com o tamanho da janela utilizada: janelas muito curtas capturam apenas variações locais e podem perder informação de sazonalidade, enquanto janelas muito longas podem aumentar a complexidade do problema e tornar o treinamento mais difícil. A comparação sistemática entre diferentes janelas permite identificar faixas em que há equilíbrio entre complexidade e capacidade de generalização.

Os modelos GRU e LSTM, avaliados isoladamente, fornecem previsões consistentes e conseguem acompanhar as tendências principais da série de casos estimados. No entanto, a combinação das previsões por meio de ensembles mostra-se uma alternativa importante para reduzir variabilidade e estabilizar o erro, já que a média das previsões de modelos distintos tende a suavizar ruídos específicos de cada arquitetura. O ensemble ponderado, em particular, utiliza pesos inversamente proporcionais ao RMSE de cada rede, atribuindo maior influência ao modelo com melhor desempenho em cada cenário.

Os gráficos gerados ao final dos experimentos ilustram a comparação entre os valores reais de casos estimados e as previsões produzidas por GRU, LSTM, ensemble médio e ensemble ponderado para a última janela analisada. Visualmente, nota-se que os modelos conseguem acompanhar a tendência e parte das oscilações da série, ainda que picos epidêmicos mais extremos sejam naturalmente mais difíceis de prever com exatidão. A combinação entre análise visual e métricas numéricas fornece um quadro completo do comportamento dos modelos na tarefa de previsão semanal de casos estimados de dengue.

Discussão e Conclusão

A modelagem desenvolvida demonstra que redes neurais recorrentes são ferramentas adequadas para capturar a dinâmica temporal dos casos estimados de dengue quando combinadas a um pipeline robusto de pré-processamento, engenharia de atributos e criação de janelas temporais. O uso de lags, médias móveis, variáveis climáticas e componentes sazonais enriqueceu a base de dados e permitiu que os modelos acessassem informações relevantes sobre dependências de curto, médio e longo prazo.

A estratégia de testar múltiplos tamanhos de janelas temporais e comparar diferentes arquiteturas (GRU e LSTM) contribuiu para uma compreensão mais aprofundada do problema e dos limites de cada abordagem. A adoção de ensembles, tanto por média simples quanto por ponderação baseada em erro, está alinhada às melhores práticas de aprendizado de máquina e tende a melhorar a estabilidade das previsões, reduzindo a sensibilidade a variações específicas de um único modelo.

Entre as principais limitações do trabalho, destacam-se a ausência de variáveis externas adicionais, como indicadores de mobilidade, fatores socioeconômicos e dados de vigilância vetorial, que poderiam refinar ainda mais as previsões; e a dificuldade natural em prever picos epidêmicos muito abruptos, característica inerente a séries epidemiológicas. Além disso, o custo computacional associado ao tuning de hiperparâmetros e ao treinamento de redes profundas exige ambiente adequado e pode ser um fator limitante em cenários de produção.

Apesar dessas limitações, o objetivo do projeto foi alcançado: foi possível desenvolver um sistema preditivo baseado em séries temporais, com dados reais do InfoDengue, que gera previsões semanais de casos estimados e fornece subsídios para a tomada de decisão em vigilância epidemiológica. Como trabalhos futuros, destacam-se a inclusão de novas fontes de dados, a experimentação com arquiteturas híbridas (como CNN-LSTM ou Transformers) e a integração do modelo em um dashboard interativo voltado a gestores de saúde.
```python
# -*- coding: utf-8 -*-
"""
Experimentos completos: windows, LSTM, GRU, KerasTuner, Ensemble, métricas.
Cole em Jupyter / Colab e rode.
"""

# 0) Instalar dependências (no Colab)
!pip install -q keras-tuner pmdarima

# ==================================================
# 1) IMPORTS
# ==================================================
import os, random, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, GRU, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# Keras Tuner
import keras_tuner as kt

# ==================================================
# 2) CONFIGURAÇÕES GLOBAIS
# ==================================================
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)

csv_url = 'https://raw.githubusercontent.com/CamillaSemenzzatt/Projeto-IV-/main/dengue_52-48.csv'
date_col = 'data_iniSE'
target_col = 'casos_est'
MODEL_DIR = 'models_all'
os.makedirs(MODEL_DIR, exist_ok=True)

# Experiment settings
WINDOWS = [52]          # só 1 janela
RUN_TUNER = False       # não roda o KerasTuner
EPOCHS = 40             # menos épocas
BATCH_SIZE = 32


# Toggle heavy steps
RUN_TUNER = True                 # Se False, usa hiperparams default
TUNER_TRIALS = 20                # número de trials
TUNER_EPOCHS = 50                # epochs por trial

# ==================================================
# 3) FUNÇÕES ÚTEIS
# ==================================================
def load_and_preprocess(csv_url, date_col='data_iniSE', target_col='casos_est'):
    df = pd.read_csv(csv_url)
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.sort_values(date_col).reset_index(drop=True)
    # Remover colunas não utilizadas
    drop_cols = ['casprov_est','casprov_est_min','casprov_est_max','casconf']
    for c in drop_cols:
        if c in df.columns:
            df.drop(columns=c, inplace=True)
    # Interpolação de dados numéricos
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    df[num_cols] = df[num_cols].interpolate(method='linear', limit_direction='both')
    # preencher tweets se houver
    if 'tweet' in df.columns:
        df['tweet'] = df['tweet'].fillna(df['tweet'].mean())
    df.set_index(date_col, inplace=True)
    df.sort_index(inplace=True)
    return df

def add_features(df, target_col='casos_est', max_lag=12, ma_windows=[7,14,21,30,52]):
    # lags do target
    for lag in range(1, max_lag+1):
        df[f'{target_col}_lag_{lag}'] = df[target_col].shift(lag)
    # médias móveis
    for w in ma_windows:
        df[f'{target_col}_ma_{w}'] = df[target_col].rolling(window=w, min_periods=1).mean()
    # amplitude térmica
    df['amplitude_termica'] = df['tempmax'] - df['tempmin'] if 'tempmax' in df.columns and 'tempmin' in df.columns else 0.0
    # features cíclicas
    df['month'] = df.index.month
    df['weekofyear'] = df.index.isocalendar().week.astype(int)
    df['month_sin'] = np.sin(2*np.pi*df['month']/12)
    df['month_cos'] = np.cos(2*np.pi*df['month']/12)
    df['week_sin'] = np.sin(2*np.pi*df['weekofyear']/52)
    df['week_cos'] = np.cos(2*np.pi*df['weekofyear']/52)
    # log-transform target
    df['target_log'] = np.log1p(df[target_col])
    df.dropna(inplace=True)
    return df

def create_windows(df, feature_cols, target_col_log='target_log', n_steps_in=52, n_steps_out=1):
    arr = df[feature_cols].values
    t = df[target_col_log].values
    X, y = [], []
    max_start = len(df) - (n_steps_in + n_steps_out) + 1
    for start in range(max_start):
        end_x = start + n_steps_in
        end_y = end_x + n_steps_out
        X.append(arr[start:end_x, :])
        y.append(t[end_x:end_y])
    X = np.array(X)
    y = np.array(y)
    if n_steps_out == 1:
        y = y.reshape(-1)
    return X, y

def train_test_split_time(X, y, train_frac=0.85):
    split = int(len(X) * train_frac)
    return X[:split], X[split:], y[:split], y[split:]

def scale_train_test(X_train, X_test, y_train, y_test):
    ns, ts, nf = X_train.shape
    scaler_X = StandardScaler().fit(X_train.reshape(-1, nf))
    X_train_s = scaler_X.transform(X_train.reshape(-1, nf)).reshape(ns, ts, nf)
    X_test_s = scaler_X.transform(X_test.reshape(-1, nf)).reshape(X_test.shape[0], ts, nf)
    scaler_y = StandardScaler().fit(y_train.reshape(-1,1))
    y_train_s = scaler_y.transform(y_train.reshape(-1,1)).reshape(-1)
    y_test_s = scaler_y.transform(y_test.reshape(-1,1)).reshape(-1)
    return X_train_s, X_test_s, y_train_s, y_test_s, scaler_X, scaler_y

def metrics_original(y_true_log, y_pred_log, scaler_y):
    y_true = np.expm1(y_true_log)
    y_pred = np.expm1(y_pred_log)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    return mse, rmse, mae, y_true, y_pred

# ==================================================
# 4) FUNÇÕES PARA CRIAR MODELOS
# ==================================================
def build_gru_tunable(hp, n_timesteps, n_features):
    units1 = hp.Int('gru_units1', 32, 256, step=32, default=128)
    units2 = hp.Int('gru_units2', 16, 128, step=16, default=64)
    dropout = hp.Float('dropout', 0.0, 0.4, step=0.05, default=0.15)
    lr = hp.Choice('lr', [1e-2, 1e-3, 1e-4], default=1e-3)
    model = Sequential([
        Input(shape=(n_timesteps, n_features)),
        GRU(units1, return_sequences=True),
        Dropout(dropout),
        GRU(units2),
        Dropout(dropout),
        Dense(hp.Int('dense', 16, 64, step=16, default=32), activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss='mse')
    return model

def build_lstm_tunable(hp, n_timesteps, n_features):
    units1 = hp.Int('lstm_units1', 32, 256, step=32, default=128)
    units2 = hp.Int('lstm_units2', 16, 128, step=16, default=64)
    dropout = hp.Float('dropout', 0.0, 0.4, step=0.05, default=0.15)
    lr = hp.Choice('lr', [1e-2, 1e-3, 1e-4], default=1e-3)
    model = Sequential([
        Input(shape=(n_timesteps, n_features)),
        LSTM(units1, return_sequences=True),
        Dropout(dropout),
        LSTM(units2),
        Dropout(dropout),
        Dense(hp.Int('dense', 16, 64, step=16, default=32), activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss='mse')
    return model

# ==================================================
# 5) FUNÇÃO DE TREINO
# ==================================================
def train_model(model, X_train, y_train, X_val, y_val, model_path, epochs=EPOCHS, batch_size=BATCH_SIZE):
    es = EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True, verbose=1)
    rlr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, verbose=1)
    ckpt = ModelCheckpoint(model_path, monitor='val_loss', save_best_only=True, verbose=1)
    hist = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                     epochs=epochs, batch_size=batch_size, callbacks=[es, rlr, ckpt], verbose=2)
    return hist

# ==================================================
# 6) FUNÇÃO PARA EXPERIMENTO COMPLETO
# ==================================================
def run_experiment_for_window(df, feature_cols, window, run_tuner=True, tuner_trials=TUNER_TRIALS):
    print("\n==========================")
    print("WINDOW:", window)
    print("========================")

    X, y = create_windows(df, feature_cols, target_col_log='target_log', n_steps_in=window)
    print("Total samples (after windows):", len(X))

    X_train, X_test, y_train, y_test = train_test_split_time(X, y, TRAIN_FRAC)
    X_train_s, X_test_s, y_train_s, y_test_s, scaler_X, scaler_y = scale_train_test(X_train, X_test, y_train, y_test)
    n_timesteps, n_features = X_train_s.shape[1], X_train_s.shape[2]

    results_local = {}

    # ---------- GRU
    if run_tuner:
        print("Rodando KerasTuner GRU")
        tuner = kt.RandomSearch(lambda hp: build_gru_tunable(hp, n_timesteps, n_features),
                                objective='val_loss', max_trials=tuner_trials,
                                executions_per_trial=1, directory='kt_gru', project_name=f'gru_w{window}')
        tuner.search(X_train_s, y_train_s, epochs=TUNER_EPOCHS, validation_data=(X_test_s, y_test_s))
        best_hp = tuner.get_best_hyperparameters(1)[0]
        model_gru = build_gru_tunable(best_hp, n_timesteps, n_features)
    else:
        model_gru = build_gru_tunable(kt.HyperParameters(), n_timesteps, n_features)

    path_gru = os.path.join(MODEL_DIR, f'gru_w{window}.keras')
    train_model(model_gru, X_train_s, y_train_s, X_test_s, y_test_s, path_gru)
    y_pred_gru_s = model_gru.predict(X_test_s).reshape(-1,1)
    y_pred_gru_log = scaler_y.inverse_transform(y_pred_gru_s).reshape(-1)
    y_test_log_inv = scaler_y.inverse_transform(y_test_s.reshape(-1,1)).reshape(-1)
    mse_g, rmse_g, mae_g, y_true_g, y_pred_g = metrics_original(y_test_log_inv, y_pred_gru_log, scaler_y)
    results_local['GRU'] = {'mse': mse_g, 'rmse': rmse_g, 'mae': mae_g, 'y_true': y_true_g, 'y_pred': y_pred_g}

    # ---------- LSTM
    if run_tuner:
        print("Rodando KerasTuner LSTM")
        tuner = kt.RandomSearch(lambda hp: build_lstm_tunable(hp, n_timesteps, n_features),
                                objective='val_loss', max_trials=tuner_trials,
                                executions_per_trial=1, directory='kt_lstm', project_name=f'lstm_w{window}')
        tuner.search(X_train_s, y_train_s, epochs=TUNER_EPOCHS, validation_data=(X_test_s, y_test_s))
        best_hp = tuner.get_best_hyperparameters(1)[0]
        model_lstm = build_lstm_tunable(best_hp, n_timesteps, n_features)
    else:
        model_lstm = build_lstm_tunable(kt.HyperParameters(), n_timesteps, n_features)

    path_lstm = os.path.join(MODEL_DIR, f'lstm_w{window}.keras')
    train_model(model_lstm, X_train_s, y_train_s, X_test_s, y_test_s, path_lstm)
    y_pred_lstm_s = model_lstm.predict(X_test_s).reshape(-1,1)
    y_pred_lstm_log = scaler_y.inverse_transform(y_pred_lstm_s).reshape(-1)
    mse_l, rmse_l, mae_l, y_true_l, y_pred_l = metrics_original(y_test_log_inv, y_pred_lstm_log, scaler_y)
    results_local['LSTM'] = {'mse': mse_l, 'rmse': rmse_l, 'mae': mae_l, 'y_true': y_true_l, 'y_pred': y_pred_l}

    # ---------- Ensemble médio
    y_pred_avg = (y_pred_g + y_pred_l) / 2
    mse_e, rmse_e, mae_e = mean_squared_error(y_true_g, y_pred_avg), np.sqrt(mean_squared_error(y_true_g, y_pred_avg)), mean_absolute_error(y_true_g, y_pred_avg)
    results_local['ENSMEAN'] = {'mse': mse_e, 'rmse': rmse_e, 'mae': mae_e, 'y_true': y_true_g, 'y_pred': y_pred_avg}

    # ---------- Ensemble ponderado (peso = inverso RMSE)
    w_g, w_l = 1/(rmse_g+1e-9), 1/(rmse_l+1e-9)
    y_pred_w = (y_pred_g*w_g + y_pred_l*w_l)/(w_g + w_l)
    mse_ew, rmse_ew, mae_ew = mean_squared_error(y_true_g, y_pred_w), np.sqrt(mean_squared_error(y_true_g, y_pred_w)), mean_absolute_error(y_true_g, y_pred_w)
    results_local['ENSW'] = {'mse': mse_ew, 'rmse': rmse_ew, 'mae': mae_ew, 'y_true': y_true_g, 'y_pred': y_pred_w}

    # Salvar previsões
    idx = df.index[-len(y_true_g):]
    outdf = pd.DataFrame({'y_true': y_true_g, 'pred_gru': y_pred_g, 'pred_lstm': y_pred_l,
                          'pred_avg': y_pred_avg, 'pred_w': y_pred_w}, index=idx)
    outdf.to_csv(os.path.join(MODEL_DIR, f'preds_w{window}.csv'))

    return results_local

# ==================================================
# 7) EXECUTAR EXPERIMENTOS
# ==================================================
df_raw = load_and_preprocess(csv_url, date_col=date_col, target_col=target_col)
df_feat = add_features(df_raw, target_col=target_col, max_lag=12, ma_windows=[7,14,21,30,52])

# Features finais
climatic_cols = [c for c in ['tempmed','umidmed','precip','tweet','Rt'] if c in df_feat.columns]
feature_cols = [f'{target_col}_lag_{lag}' for lag in range(1,13)]
feature_cols += [f'{target_col}_ma_{w}' for w in [7,14,21,30,52]]
feature_cols += ['amplitude_termica','month_sin','month_cos','week_sin','week_cos']
feature_cols += climatic_cols
feature_cols = [c for c in feature_cols if c in df_feat.columns]

print("Features finais:", feature_cols)

all_results = {}
start_total = time.time()
for w in WINDOWS:
    res = run_experiment_for_window(df_feat, feature_cols, w, run_tuner=RUN_TUNER, tuner_trials=TUNER_TRIALS)
    all_results[w] = res
end_total = time.time()
print("Tempo total (s):", end_total - start_total)

# ==================================================
# 8) AGREGAR E EXIBIR RESULTADOS
# ==================================================
rows = []
for w, r in all_results.items():
    for mname, metrics in r.items():
        rows.append({'window': w, 'model': mname, 'mse': metrics['mse'], 'rmse': metrics['rmse'], 'mae': metrics['mae']})
res_df = pd.DataFrame(rows).sort_values(['window','rmse']).reset_index(drop=True)
display(res_df)

# ==================================================
# 9) PLOTS COMPARATIVOS (última janela)
# ==================================================
last_w = WINDOWS[-1]
preds_path = os.path.join(MODEL_DIR, f'preds_w{last_w}.csv')
if os.path.exists(preds_path):
    preds = pd.read_csv(preds_path, index_col=0, parse_dates=True)
    plt.figure(figsize=(12,4))
    plt.plot(preds['y_true'], label='real')
    plt.plot(preds['pred_gru'], label='gru')
    plt.plot(preds['pred_lstm'], label='lstm')
    plt.plot(preds['pred_avg'], label='ensemble_avg', linestyle='--')
    plt.plot(preds['pred_w'], label='ensemble_weighted', linestyle='--')
    plt.legend(); plt.title(f'Predições (window={last_w})'); plt.show()

# ==================================================
# 10) Salvar resumo
# ==================================================
res_df.to_csv(os.path.join(MODEL_DIR, 'summary_results.csv'), index=False)
print("Resumo salvo em:", os.path.join(MODEL_DIR, 'summary_results.csv'))
```
